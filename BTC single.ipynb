```python
import pandas as pd
import numpy as np
from datetime import datetime

data = pd.read_csv('Alldata_raw_delta_log_AR.csv')
data['Date'] = data['Date'].apply(lambda x: datetime.strptime(x, '%d-%b-%y'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 假設 data 是你的 DataFrame，並且已經包含時間索引
data['timestamp'] = pd.to_datetime(data['Date'])
                                   
# 設定時間索引
data.set_index('timestamp', inplace=True)

# 建立一個新的 figure
fig, ax1 = plt.subplots(figsize=(10, 5))

# 在主要軸上繪製 BTC
ax1.plot(data['Log_BTC'], label='Log_BTC', color='orange')
ax1.tick_params('y')

# 創建次要軸並在次要軸上繪製 Log_BTC
ax2 = ax1.twinx()
ax2.plot(data['BTC'], label='BTC')
ax2.tick_params('y')

# 設定標題
#plt.title('BTC & Log Return')

# 設定 x 軸日期格式
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax1.xaxis.set_major_locator(mdates.YearLocator())

# 顯示圖例
'''
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right', facecolor='none')
'''

# 設定整個 Figure 的背景色為透明
fig.patch.set_facecolor('none')
ax1.set_facecolor('none')
ax2.set_facecolor('none')

fig.tight_layout()
ax1.yaxis.tick_right()  # 將 ax1 的 y 軸刻度顯示在右邊
ax2.yaxis.tick_left()   # 將 ax2 的 y 軸刻度顯示在左邊
plt.show()
```
