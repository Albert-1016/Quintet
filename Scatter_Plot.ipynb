# 可調整限制範圍在-0.5~0.5

# Scenario 1 X_t

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取包含数据的CSV文件
data = pd.read_csv('Alldata_raw_delta_log_AR.csv')

# 获取数据中的經濟指標列名，假设它们是列名的後23列
economic_indicators = data.columns[1:24]

# 设置子图的行数和列数
num_rows = 5
num_cols = 5

# 创建一个包含20张散点图的子图布局
fig, axs = plt.subplots(num_rows, num_cols, figsize=(36, 21), facecolor='none')

# 将经济指标与报酬率绘制为散点图
for i, indicator in enumerate(economic_indicators):
    row = i // num_cols
    col = i % num_cols
    ax = axs[row, col]
    ax.scatter(data[indicator], data['rt(1)'])
    ax.set_title(indicator, fontsize=30)
    ax.set_ylim(-0.5, 0.5)
    ax.set_facecolor('none')
    
# 调整子图布局
plt.tight_layout()

# 显示散点图
plt.show()
```

# Scenario 2 D_X_t

```python
import pandas as pd
import matplotlib.pyplot as plt

# 获取数据中的經濟指標列名，假设它们是列名的後23列
economic_indicators = data.columns[27:50]

# 设置子图的行数和列数
num_rows = 5
num_cols = 5

# 创建一个包含20张散点图的子图布局
fig, axs = plt.subplots(num_rows, num_cols, figsize=(36, 21), facecolor='none')

# 将经济指标与报酬率绘制为散点图
for i, indicator in enumerate(economic_indicators):
    row = i // num_cols
    col = i % num_cols
    ax = axs[row, col]
    ax.scatter(data[indicator], data['rt(1)'])
    #ax.set_title(indicator, fontsize=30)
    coin_name = indicator.split('_')[1]  # 用_分隔然後利用index取後面
    ax.set_title(f'{coin_name}', fontsize=30)
    ax.set_ylim(-0.5, 0.5)
    ax.set_facecolor('none')

# 调整子图布局
plt.tight_layout()

# 显示散点图
plt.show()
```
